```python
def read_txt_file(files):
    data = []
    for file in files:
        with open(file,encoding='utf-8') as f:
            val = [item.strip() for item in f.readlines()]
            data.append(val)
    return data
```

```python
txt_data = read_txt_file(['data/corpus.txt'])[0][:1000]
```

```python
import string
def clean_txt(sent):
    return sent.translate(str.maketrans('', '', string.punctuation))
```

```python
from pyvi import ViTokenizer
import re
def tokenizer(sent):
    result = ViTokenizer.tokenize(sent)
    return [item.lower() for item in re.findall(r'\S+', str(result))]
```

```python
txt_data = [clean_txt(sent) for sent in txt_data]
```

```python
txt_data = [tokenizer(item) for item in txt_data]
```

```python
from gensim.models import Word2Vec
model = Word2Vec(txt_data, min_count=1)
```

```python
print(model['lịch_sử'])
```

```python
model.save('word2vec.bin')
```

```python
pretrain_model = Word2Vec.load('word2vec.bin')
```

```python
import numpy as np

def get_sentence_vector(sentence):
    result = tokenizer(clean_txt(sentence))
    _tuple = []
    for item in result:
        try:
            vector = pretrain_model[item]
            _tuple.append((item, vector))
        except:
            _tuple.append((item, np.zeros((100,), dtype=float)))
    return _tuple
```

```python
get_sentence_vector('niềm tin đã mất giọt nước mắt cuốn kí ức anh chìm sâu')
```
