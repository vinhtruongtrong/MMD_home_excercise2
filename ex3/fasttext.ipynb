```python
from gensim.models.wrappers import FastText
embedding_model = FastText.load_fasttext_format('embedding/cc.vi.300.bin')
```

```python
import re
def embedding(sent):
    output = []
    query = re.findall(r'\S+', str(sent))
    for word in query:
        try:
            output.append((word, embedding_model[word.toLower()]))
        except:
            output.append((word, embedding_model['unknown']))
    return output
```

```python
from pyvi import ViTokenizer
def tokenize(sent):
    return ViTokenizer.tokenize(sent)

def get_vector_sentence(sent):
    return embedding(tokenize(sent))
```

```python
vec = get_vector_sentence('tìm về nơi đâu đôi chân anh lạc trôi giữa bầu trời')
```
